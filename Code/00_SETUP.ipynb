```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 1 — Mount Google Drive
# ─────────────────────────────────────────────────────────────────────────────
# What this does:
#   Connects Google Colab to your Google Drive so we can read and write files.
#   A popup will appear asking you to sign in — that is normal and expected.
#   Click the link, choose your Google account, and copy the code back here.
# ─────────────────────────────────────────────────────────────────────────────

from google.colab import drive
drive.mount('/content/drive')

print('  Google Drive mounted successfully.')
print('   Your files are accessible at: /content/drive/MyDrive/')
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 2 — Install all required Python packages
# ─────────────────────────────────────────────────────────────────────────────
# What this does:
#   Installs the Python libraries (tools) that the rest of the pipeline needs.
#   Google Colab already has pandas, numpy, matplotlib pre-installed.
#   We only need to install the finance-specific ones.
#
# This cell will take about 60-90 seconds. You will see a lot of green text
# scrolling — that is normal. Wait for the '✅ All packages installed' message.
# ─────────────────────────────────────────────────────────────────────────────

import subprocess, sys

PACKAGES = [
    ('yfinance',     'Downloads stock market price data from Yahoo Finance'),
    ('pytrends',     'Downloads Google Search Trends (SVI data)'),
    ('xgboost',      'XGBoost machine learning model'),
    ('shap',         'SHAP — explains what each feature contributes to predictions'),
    ('statsmodels',  'ADF stationarity test for time series'),
    ('requests',     'Makes HTTP requests (used for Fear & Greed API)'),
]

print('Installing packages...\n')
all_ok = True
for pkg, description in PACKAGES:
    result = subprocess.run(
        [sys.executable, '-m', 'pip', 'install', pkg, '-q'],
        capture_output=True, text=True
    )
    if result.returncode == 0:
        print(f'{pkg:15}  {description}')
    else:
        print(f'{pkg:15}  FAILED — {result.stderr[:80]}')
        all_ok = False

print()
if all_ok:
    print('All packages installed successfully.')
else:
    print('⚠️  Some packages failed. Try re-running this cell.')
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 3 — Define the base path and create all folders
# ─────────────────────────────────────────────────────────────────────────────
# What this does:
#   Creates the folder structure inside your Google Drive.
#   If the folders already exist, it skips them (safe to re-run).
# ─────────────────────────────────────────────────────────────────────────────

import os

# This is the root folder for your entire capstone project
BASE = '/content/drive/MyDrive/CapstoneDA'

FOLDERS = [
    f'{BASE}/data/raw',       # Raw downloads — never modified after saving
    f'{BASE}/data/clean',     # Cleaned, aligned data
    f'{BASE}/data/features',  # Engineered features (RSI, MACD, etc.)
    f'{BASE}/data/selected',  # Features chosen by LASSO + RFE
    f'{BASE}/data/results',   # Model outputs, hypothesis test results
    f'{BASE}/data/plots',     # All charts and visualizations
]

print(f'Creating folder structure inside: {BASE}\n')
for folder in FOLDERS:
    os.makedirs(folder, exist_ok=True)
    # Show only the part after CapstoneDA/ for readability
    short = folder.replace(BASE, 'CapstoneDA')
    print(f'  ✅  {short}/')

print()
print('All folders ready inside your Google Drive.')
```
